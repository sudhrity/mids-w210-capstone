```python
#
# function to run a select query and return rows in a pandas dataframe
# pandas puts all numeric values from postgres to float
# if it will fit in an integer, change it to integer
#

def my_select_query_pandas(query, rollback_before_flag, rollback_after_flag):
    "function to run a select query and return rows in a pandas dataframe"
    
    if rollback_before_flag:
        connection.rollback()
    
    df = pd.read_sql_query(query, connection)
    
    if rollback_after_flag:
        connection.rollback()
    
    # fix the float columns that really should be integers
    
    for column in df:
    
        if df[column].dtype == "float64":

            fraction_flag = False

            for value in df[column].values:
                
                if not np.isnan(value):
                    if value - math.floor(value) != 0:
                        fraction_flag = True

            if not fraction_flag:
                df[column] = df[column].astype('Int64')
    
    return(df)
```

```python
sr = SearchEngine()

def update_zipcode_details(year, zipcode):

    #print('Zipcode=', zipcode)
    
    connection.rollback()
    
    z = sr.by_zipcode(zipcode)
    
    updated_rows = 0
    
    try:
        
        if z is not None:
        
            city = z.major_city
            state = z.state_abbr
            county = z.county
            median_income = z.median_household_income

            print(zipcode, state, county, city, median_income)

            query = '''

            UPDATE zipcode_detail 
                    SET state = %s, 
                        county = %s,
                        city_neighborhood = %s, 
                        year = %s,
                        median_income = %s
                        WHERE zipcode = %s;
                        '''



            cursor.execute(query, (state, county, city, year, median_income, zipcode))

            updated_rows = cursor.rowcount
        
    except (Exception, psycopg2.DatabaseError) as error:
        print(error)
    
    finally:
        if connection is not None:
                connection.commit()
        
    connection.commit()
    
    return updated_rows
```
