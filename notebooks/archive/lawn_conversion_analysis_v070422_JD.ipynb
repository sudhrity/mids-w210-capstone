```python
lc = base_lc
```

```python
# creates simple variables for vegetation(trees & grass) and impervious(soil & impervious)
lc['vegetation'] = lc['vegetation_trees_area']+lc['vegetation_grass_area']
lc['impervious'] = lc['impervious_area']+lc['soil_area']
lc['ext_impervious'] = lc['impervious_area']+lc['soil_area']+lc['turf_area']
```

```python
# creates percentages for each class
lc['water_%']=lc['water_area']/lc['total_area']*100
lc['trees_%']=lc['vegetation_trees_area']/lc['total_area']*100
lc['grass_%']=lc['vegetation_grass_area']/lc['total_area']*100
lc['turf_%']=lc['turf_area']/lc['total_area']*100
lc['impervious_%']=lc['impervious']/lc['total_area']*100
lc['vegetation_%']=lc['vegetation']/lc['total_area']*100
lc['ext_impervious_%']=lc['ext_impervious']/lc['total_area']*100
```

```python
lc_1 = lc
```

```python
def diff_column(df,column_name):
    '''Calculates delta from the consecutive 2 year period for a zipcode and a class'''
    df[column_name + '_delta'] = df[column_name] - df.groupby(['polygon'])[column_name].shift()
    return df
```

```python
feature_list = ['water_area',
                'vegetation_trees_area',
                'vegetation_grass_area',
                'turf_area',
                'impervious_area',
                'soil_area',
                'vegetation',
                'impervious',
                'water_%',
                'trees_%',
                'grass_%',
                'turf_%',
                'impervious_%',
                'vegetation_%',
                'ext_impervious_%']
for feature in feature_list:
    diff_column(lc_1, feature)
```
